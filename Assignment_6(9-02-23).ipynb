```python
class vehicle:
    # constructor
    def __init__(self,name_of_vehicle,max_speed,average_of_vehicle):
        self.name_of_vehicle = name_of_vehicle
        self.max_speed = max_speed
        self.average_of_vehicle = average_of_vehicle
```

```python
obj_vehicle = vehicle('BMW',210,24)  # instance of 'vehicle' class
```

```python
print(obj_vehicle.name_of_vehicle)
print(obj_vehicle.max_speed)
print(obj_vehicle.average_of_vehicle)
```

```python
class car(vehicle):  # single inheritance
    # method
    def seating_capacity(self,capacity):
        return self.name_of_vehicle,capacity
```

```python
obj_car = car('Audi',220,25)
```

```python
obj_car.seating_capacity(4)
```

Ans) **Multiple Inheritance:-** In multiple inheritance subclass inherits from **`two or more parents`**

```python
# Multiple inheritance example
class parent:
    # constuctor
    def __init__(self,father_name,mother_name):
        self.father_name = father_name
        self.mother_name = mother_name
        
        
class name:
    # method
    def parent_name(self):
        print(f'My father name is {self.father_name} and my mother name is {self.mother_name}')

class son(parent,name):
              pass
```

```python
obj_son = son('Sunil Singh','Sanju Singh') # instance of 'son' class
```

```python
obj_son.parent_name()
```

Ans) In Python, `getter` and `setter` methods are used to `access` and `modify` the ***private variable of a class*** respectively. 

***`These methods  are often used to enforce data encapsulation, which is the practice of restricting direct access to an object's attribute from outside the object.`***

**getter** method is used to get the value of the private variable.`It is used to retrieve the value of an attribute in a class.` It is named using **get_** prefix followed by attribute name.

**setter** method is used to set the value of the private variable.`It is used to modify the value of an attribute in a class.` It is named using **set__** prefix followed by attribute name.


```python
class student:
    # constructor 
    def __init__(self,name,email,age):
        self.__student_name = name
        self.__student_email = email
        self.__student_age = age
        
    # methods
    def get_student_details(self):
        print('Student name:',self.__student_name)
        print('Student email:',self.__student_email)
        print('Student age:',self.__student_age)
        
    def get_email(self):
        return self.__student_email
    
    def set_email(self,email):
        self.__student_email = email
```

```python
obj_student = student('Shalu Singh','Singh@gmail.com',22) # instace of 'student' class
```

```python
obj_student.get_student_details()  # getter method
```

```python
obj_student.get_email() # getter method
```

```python
obj_student.set_email('Shalu@gmail.com') # setter method
```

```python
obj_student.get_student_details()
```

Ans) Method overriding in Python refers to the practice of **`defining method in a subclass that has the same name and arguments as a method in superclass`**.

When the method is called on **instance of subclass**, the `subclass method is invoked instead of the superclass method`.

```python
# method overriding
class Animal:
    # method 
    def tiger_(self):
        print('Tiger is a carnivorous')
        
class cat(Animal):
    
    # method overriding
    def tiger_(self):
        print('Cat belongs to tiger family')
```

```python
obj_cat = cat() # instance of 'cat' class
obj_cat.tiger_()
```
