**class:-** A class is a `blueprint` that defines ***the properties and behaviors of a group of object.*** It `binds the data and functions required to create and manipulate objects of that class`.

***`A class can be thought of user-defined data type that provides a way to organize and structure code.`***

**object:-** A object is a `instance` of a class. It is real entity that takes space in memory at runtime. `It has all the properties and attribute of a class` and is use to access data or perform operation.



```python
# Creating a car class having properties :- color, model
# and functions like :- start_engine, stop_engine

class car:
    def __init__(self,model,color):
        self.model = model
        self.color = color
        
        # class method
    def start_engine(self):
        print('Starting the engine')
            
    def stop_engine(self):
        print('Stoping the engine')
            
    def car_details(self):
        return self.model,self.color
```

```python
bmw = car('BMW','Red')  # instance of 'car' class
bmw.start_engine()   
bmw.stop_engine()
bmw.car_details()
```

Ans) In python **`__init__()`** function acts as a **special method** that is `automatically called when object is created` just like **`constructor`**.

It is used in classes to **`initialize the attributes of object when it is created`**. It cane be used to **set default value or to perform any necessary setup for object.**

```python
# Creating a car class having properties :- color, model
# and functions like :- start_engine, stop_engine

class car:
    def __init__(self,model,color): # acts as constuctor, initialize object attributes
        self.model = model  # objects attributes
        self.color = color
        self.max_speed = '280 km/h'
        # class method
    def start_engine(self):
        print('Starting the engine')
            
    def stop_engine(self):
        print('Stoping the engine')
            
    def car_details(self):
        return self.model,self.color,self.max_speed
```

```python
audi = car('Audi','White')  # instance of 'car' class
audi.start_engine()   
audi.stop_engine()
audi.car_details()
```

Ans) **Inheritance:-** It is one of the pillar of OOPs use for `creation of new class based on existing class`.**Inheriting all its properties and behavior while allowing for new properties and behavior to be added**. The existing class is known as **`parent`** or **`superclass`**, and the new class is called **`child`** or **`subclass`**.

***There are 5 types of inheritance:-***

1.) **Single Inheritance:-** In single inheritance, a subclass inherits properties and behavior of a **`single parent class`**



2.) **Multiple Inheritance:-** In multiple inheritance, a subclass inherits properties and behavior of **`multiple parent 
class`**



3.) **Hierarchical Inheritance:-** In hierarchical inheritance, a subclass inherits properties and behavior of **`single parent class, but the parent class itself inherits from one or more parent class`**



4.) **Multilevel Inheritance:-** In multilevel inheritance, a subclass inherits **`from a parent class, which in turn inherits form its own parent class, and so on, forming a hierarchy of classes.`**



5) **Hybrid Inheritance:-** It is a combination of `two or more types of inheritance`
