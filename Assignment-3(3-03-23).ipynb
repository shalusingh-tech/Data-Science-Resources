Function Syntax:- `def function_name(parameters):`

```python
# Function that returns list of odd number in range 1 - 25

def odd_filter():
    '''This function returns a list of odd number in (1-25) range'''
    odd_lst = [] # creating empty 
    for i in range(1,26):
        if (i % 2 != 0):  #logic for finding odd number 
            odd_lst.append(i)
    return odd_lst
    
    
    
result = odd_filter()  # calling function 
print('Print odd list of odd number values in range 1 - 25 are:')
print(result)
```

Ans) **`*args`** allows us to pass any number of arguments to the function. The arguments are then collected in `tuple` called **args** which we can `iterate`.

`**kwagrs` argument is used to pass **variable number of keyword arguments** to a function. When `**kwargs**` is used in a function definition, it collects all the extra keyword arguments that are passed to the function as **`dictionary`**. 

***These two allows you to pass any number of argument to a function without having to explicitly define them in the function signature***

```python
# function 1
def func_1(*args): # here we are using *args parameter to pass any number of arguments to the funtion
    return args

def func_2(**kwargs): # here we are using **kwargs parameter to pass variable number of keyword arguments
    return kwargs

def func_3(*args,**kwargs):
    return args,kwargs


# function call
r_func_1 = func_1(1.5,5,9,'pwskills',22,2+5j)
r_func_2 = func_2(name = 'Shalu',course = 'Data Science Masters',age = 22)
r_func_3 = func_3(2.5,8,name= 'Shalu',age = 22)

print('Func_1 return:',r_func_1)
print('Type(func_1)',type(r_func_1))
print('-----------------------\n')

print('Func_2 return:',r_func_2)
print('Type(func_2)',type(r_func_2))
print('-----------------------\n')

print('Func_2 return:',r_func_3)
print('Type(func_2)',type(r_func_3))
```

Ans) An **iterator** is an object in Python that represents **`stream of data`**. It is used to iterate through a sequence or collection of data,such as list or a tuple, and it **`returns one item at a time`**.

***In python, an iterator is an object that implements the iterator protocol, which consists of two methods `__iter__()` and `__next__()`***

**`iter()`** methods returns the iterator object itself.

**`next()`** method returns the next value in the sequence. Raises ***StopIteration*** exception when `no  more values to be returned`

```python
my_list = list(range(2,22,2)) # [2,4,6,8,10,12,14,1,6,18,20]
my_iterator = iter(my_list)  # returns the iter object

# print first five elements of list
for i in range(5):
    print(next(my_iterator)) # return next value in sequence
```

Ans) In python, a generator function is a type of function that can be used to **`create iterator`**. Unlike regular functions, which return a value and terminate, ***a generator function can `yield` value and then pause execution, allowing the function to be resumed from where it left off when the next value is requested.***

**`yield`** keyword is used in a generator function to indicate ***where the function should `pause` and `return` a value to the the caller.*** The value that is yielded can be any Python object such as `a number, a string, a list or even another generator function`.

**NOTE:-** Generator function uses memory in more efficient way 

```python
# creating generator function to print first n natural number

def n_natural_number(n):
    '''This is a generator function which yield first n natural number'''
    for i in range(1,n+1):
        yield i # function pause and returns value to caller
```

```python
n = int(input('Enter the input: '))
n_natural_number_gen = n_natural_number(n)
type(n_natural_number_gen)
```

```python
print(f'1st {n} natural numbers are:')
for i in n_natural_number_gen:
    print(i)
    
```
