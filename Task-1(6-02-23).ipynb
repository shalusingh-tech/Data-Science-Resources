```python
from functools import reduce
def flat_list_product(lst):
    '''This function flatend the list that only contains numeric elements'''
    
    flat_lst = [] # empty list
    
    for i in lst: # itterating trough main list
        i_type = type(i) # storing type of element
        if(i_type == int or i_type == float): # checking if element in int or float
            flat_lst.append(i)
        elif (i_type == bool):  # internally bool is also treated as number (True = 1,False = 0)
            flat_lst.append(i)
        elif (i_type == list or i_type == tuple or i_type == set): # check if element is another sequence data type
            for j in i:
                j_type = type(j)
                if(j_type == int or j_type == float or j_type == bool):
                    flat_lst.append(j)
                    
        elif (i_type == dict):  # is element is dictonary
            for k in i:
                if (type(k) == int or type(k)==float): # is key numeric
                    flat_lst.append(k)
                    
                if (type(i[k]) == int or type(i[k]) == float or type(i[k]) == bool): # check value is numeric
                    flat_lst.append(i[k])
                elif (type(i[k]) == list or type(i[k]) == tuple or type(i[k]) == set): # check value is sequence
                    for j in i[k]:
                        j_type = type(j)
                        if(j_type == int or j_type == float or j_type == bool):
                            flat_lst.append(j)
                   
    product = reduce(lambda x,y: x*y,flat_lst) # product of all elements in flat list
    return product,list(flat_lst)
```

```python
list1 = [1,2,3,4, [44,55,66, True],False, (34,56,78,89,34), {1,2,3,3,2,1}, {1:34, "key2": [55, 67, 78, 89], 4: (45,
22, 61, 34)}, [56, 'data science'], 'Machine Learning']

result,f_list = flat_list_product(list1) # function call
print('Flatend list is:',f_list)
print('Product result =',result)
```

```python
def msg_encryption_func(msg):
    '''This function encrypt the input message'''
    msg =  msg.lower() # all charcters will change to lower case 
    encrypted_msg = [] # empty list
    # chr():- it converts ascii value to its coresponding character
    encryp_map = {chr(x):chr(122+(97-x)) for x in range(97,123)} # dictonary comprehension
    encryp_map[' '] = '$' # whitespace --> $
    encryp_map['.'] = '.'  # . --> .

    for i in msg:
        encrypted_msg.append(encryp_map[i])  # mapping to new encryted value and appending to list
        
    encrypted_msg = "".join(encrypted_msg)  
    return encrypted_msg    
```

```python
msg = "I want to become a Data Scientist."
encryp_msg = msg_encryption_func(msg)
print('Original message:',msg)
print('Encrypted message:',encryp_msg)
```
