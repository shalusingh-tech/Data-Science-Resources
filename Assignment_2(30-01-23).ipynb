```python
# Program to calculate Grade base on percentage
# INPUT = Percentage
# OUTPUT = Grade

def check_grade(percentage):
    '''This function takes percentage as input and calcutes grade based on it'''
    
    if (percentage > 90):
        return 'A'
    elif (percentage > 80 and percentage <= 90):
        return 'B'
    elif (percentage >= 60 and percentage <= 80):
        return 'C'
    else:
        return 'D'
```

```python
percentage = float(input('Enter the percentage you obtained:'))
if(percentage > 0):
    grade = check_grade(percentage)
    print('Your grade is',grade)
else:
    print('Sorry! Wrong input')
```

```python
# Program to compute Tax to be paid
# INPUT = Cost Price(Rs)
# OUTPUT = Tax

def tax_calc(price):
    '''This function calculates the road Tax based on cost price'''
    
    if (price > 1e5):
        return 15
    elif (price > 5e4 and price <= 1e5):
        return 10
    else:
        return 5
    
```

```python
cost_price = float(input('Enter cost price of bike Rs:'))

if (cost_price >0):
    tax = tax_calc(cost_price)
    print(f'Tax to be paid = {tax}%')
else:
    print('Sorry! Wrong Input')
```

```python
# Program to display famous monument of the city
# INPUT = City
# OUTPUT = monument

def famous_monument(city):
    '''This function takes city as input and display famous monument of the city '''
    
    if (city == 'Delhi'):
        return "Red Fort"
    elif (city == "Agra"):
        return "Taj Mahal"
    elif (city == 'Jaipur'):
        return "Jal Mahal"
```

```python
city = input('Enter city:')

monument = famous_monument(city)
print(f'Famous monument of {city} city is {monument}')
```

```python
# Count program
def count_div_3(num):
    '''This function count how many times a given number can be divided by 3 before it is less than or equal to 10'''
    
    count = 0
    while(num>=10):
        count += 1
        num /= 3
    return count
```

```python
num = float(input('Enter a number: '))

count = count_div_3(num)
print("Count = ",count)
```

```python
# Pattern -1 
def pattern_1(row):
    i = 1 
    while (i <= row):
        j = i
        while (j <= i):
            print("*"*j,end = "")
            j+= 1
        print()
        i+= 1
        
    
```

```python
# Pattern-2

def pattern_2(row):
    i = 1 
    while(i<=row):
        j = i
        while(j<=i):
            print(" "*(row-j),end='')
            print('*'*j,end='')
            j+=1
        print()
        i+=1
        
           
    
```

```python
# pattern_3
def pattern_3(row):
    i = 1
    while(i <= row):
        j = i
        while(j<=i):
            print(" "*(row-j),end = "")
            print("*"*((2*j)-1),end="")
            j+=1
        print()
        i+=1
        
```

```python
# Displaying 3 different patterns
row = int(input('Enter number of rows: '))
print("Pattern 1")
pattern_1(row)
print("Pattern 2")
pattern_2(row)
print("Pattern 3")
pattern_3(row)
```
